```python
from xml.etree import ElementTree
from datetime import datetime
```

```python
def split(word): 
    return [char for char in word]  

alphabet = split("abcdefghijklmnopqrstuvwxyz*")
```

```python
query = "pes_stats_"
suggest_url = "http://suggestqueries.google.com/complete/search?output=toolbar&hl=fr&q="+query
suggest_list = []
for i in alphabet:
    response = requests.get(suggest_url+i)
    print(response.content)
    tree = ElementTree.fromstring(response.content.decode("utf-8", "backslashreplace"))
    for node in tree:
        suggest_list.append(node[0].attrib["data"])
```
